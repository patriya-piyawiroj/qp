```python
import pandas as pd
import numpy as np
import yfinance as yf
import statsmodels.api as sm
from statsmodels.tsa.ar_model import AutoReg
from scipy.stats import chi2

def get_data(ticker, start_date, end_date):
    data = yf.download(ticker, start=start_date, end=end_date)
    return data['Adj Close']

def compute_returns(prices):
    returns = np.log(prices).diff().dropna()
    return returns

def fit_mean_model(returns):
    model = AutoReg(returns, lags=1).fit()
    residuals = model.resid
    return residuals

def compute_lm_statistic(residuals, lags):
    squared_residuals = residuals**2
    T = len(squared_residuals)
    lagged_residuals = sm.add_constant(sm.tsa.lagmat(squared_residuals, lags))
    aux_reg = sm.OLS(squared_residuals[lags:], lagged_residuals[lags:]).fit()
    LM_stat = T * aux_reg.rsquared
    return LM_stat

def perform_arch_test(ticker, start_date, end_date, lags_list):
    prices = get_data(ticker, start_date, end_date)
    returns = compute_returns(prices)
    residuals = fit_mean_model(returns)
    
    results = {}
    for lags in lags_list:
        LM_stat = compute_lm_statistic(residuals, lags)
        p_value = 1 - chi2.cdf(LM_stat, lags)
        results[lags] = {'LM_stat': LM_stat, 'p_value': p_value}
    
    return results

# Perform ARCH test for each ticker
# Parameters
tickers = ['AAPL', '^SPX']
lags_list = [3, 6]
start_date = '2005-01-04'
end_date = '2022-12-30'
results = {}
for ticker in tickers:
    results[ticker] = perform_arch_test(ticker, start_date, end_date, lags_list)

print(results)
```

```python
from arch import arch_model
import matplotlib.pyplot as plt


def fit_arch_model(returns, lags):
    model = arch_model(returns, vol='ARCH', p=lags)
    fitted_model = model.fit(disp='off')
    return fitted_model
def sum_arch_coefficients(fitted_model):
    params = fitted_model.params
    return params
def plot_volatility_and_returns(returns, conditional_volatility, ticker):
    plt.figure(figsize=(14, 7))
    plt.plot(returns.index, np.abs(returns), label='Absolute Returns', color='blue')
    plt.plot(returns.index, conditional_volatility, label='Conditional Volatility', color='red')
    plt.title(f'Conditional Volatility vs Absolute Returns for {ticker}')
    plt.legend()
    plt.show()

def estimate_unconditional_volatility(fitted_model):
    params = fitted_model.params
    omega = params['omega']
    alpha = params[[f'alpha[{i}]' for i in range(1, fitted_model.model.volatility.p + 1)]].sum()
    unconditional_volatility = np.sqrt(omega / (1 - alpha))
    return unconditional_volatility

lags=3
results = {}
for ticker in tickers:
    prices = get_data(ticker, start_date, end_date)
    returns = compute_returns(prices)
    returns = returns*100
    fitted_model = fit_arch_model(returns, lags)
    arch_coeff = sum_arch_coefficients(fitted_model)
    
    conditional_volatility = fitted_model.conditional_volatility
    unconditional_volatility = estimate_unconditional_volatility(fitted_model)
    sample_std_dev = returns.std()

    # Store results
    results[ticker] = {
        'Unconditional_Volatility': unconditional_volatility,
        'Sample_Standard_Deviation': sample_std_dev,
        'Model_Summary': fitted_model.summary(),
        'ARCH_sum': sum(arch_coeff),
        'ARCH_coefficients': arch_coeff
    }
    
    plot_volatility_and_returns(returns, conditional_volatility, ticker)


for ticker, result in results.items():
    print(f"Ticker: {ticker}")
    print(f"ARCH coefficients: {results[ticker]['ARCH_coefficients']}\n")
    print(f"Sum of ARCH coefficients: {results[ticker]['ARCH_sum']}\n")
    print(results[ticker]['Model_Summary'])
```

```python
def sum_arch_and_garch_coefficients(fitted_model):
    params = fitted_model.params
    arch_params = params[[f'alpha[{i}]' for i in range(1, fitted_model.model.volatility.p + 1)]]
    garch_params = params[[f'beta[{i}]' for i in range(1, fitted_model.model.volatility.q + 1)]]
    arch_sum = arch_params.sum()
    garch_sum = garch_params.sum()
    return arch_sum, garch_sum


results = {}
for ticker in tickers:
    prices = get_data(ticker, start_date, end_date)
    returns = compute_returns(prices)*100
    model = arch_model(returns, vol = 'GARCH', p = 1, q = 1)
    fitted_model = model.fit(disp = 'off', show_warning = False)
    
    conditional_volatility = fitted_model.conditional_volatility
    arch_sum, garch_sum = sum_arch_and_garch_coefficients(fitted_model)
    unconditional_volatility = fitted_model.params['omega'] / (1 - (arch_sum + garch_sum))
    sample_std_dev = returns.std()

    # Store results
    results[ticker] = {
        'ARCH_Sum': arch_sum,
        'GARCH_Sum': garch_sum,
        'Unconditional_Volatility': unconditional_volatility,
        'Sample_Standard_Deviation': sample_std_dev,
        'Model_Summary': fitted_model.summary()
    }

    # Plot conditional volatility vs absolute returns
    plot_volatility_and_returns(returns, conditional_volatility, ticker)

for ticker, result in results.items():
    print(f"Ticker: {ticker}")
    print(f"Sum of ARCH coefficients: {result['ARCH_Sum']}")
    print(f"Sum of GARCH coefficients: {result['GARCH_Sum']}")
    print(f"Unconditional Volatility (GARCH Model): {result['Unconditional_Volatility']}")
    print(f"Sample Standard Deviation of Returns: {result['Sample_Standard_Deviation']}\n")
    print(result['Model_Summary'])
```

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

def get_data(ticker, start_date, end_date):
    data = yf.download(ticker, start=start_date, end=end_date)
    return data['Adj Close']

def compute_returns(prices):
    returns = np.log(prices).diff().dropna()
    return returns

def fit_arch_model(returns, lags):
    model = arch_model(returns, vol='ARCH', p=lags)
    fitted_model = model.fit(disp='off')
    return fitted_model

def fit_garch_model(returns):
    model = arch_model(returns, vol = 'GARCH', p = 1, q = 1)
    fitted_model = model.fit(disp = 'off', show_warning = False)
    return fitted_model

def estimate_unconditional_volatility(fitted_model):
    params = fitted_model.params
    omega = params['omega']
    alpha = params[[f'alpha[{i}]' for i in range(1, fitted_model.model.volatility.p + 1)]].sum()
    return omega / (1 - alpha)

def compute_volatility_forecasts(fitted_model, unconditional_volatility, h):
    forecasts = []
    for i in range(h):
        forecast = np.sqrt(fitted_model.params['omega'] + fitted_model.params['alpha[1]'] * (unconditional_volatility ** 2))
        forecasts.append(forecast)
    return forecasts

# Parameters
h = 100  # Number of days for volatility forecasts
lags = 3

# Compute unconditional volatility for each series
unconditional_volatilities = {}
for ticker in tickers:
    prices = get_data(ticker, start_date, end_date)
    returns = compute_returns(prices)*100
    fitted_model = fit_arch_model(returns, lags)
    unconditional_volatility = estimate_unconditional_volatility(fitted_model)
    unconditional_volatilities[ticker] = unconditional_volatility

# Compute volatility forecasts for ARCH(3) and GARCH(1,1) models
volatility_forecasts = {}
for ticker in tickers:
    prices = get_data(ticker, start_date, end_date)
    returns = compute_returns(prices)*100
    arch_model = fit_arch_model(returns, lags)
    garch_model = fit_garch_model(returns)
    arch_forecasts = compute_volatility_forecasts(arch_model, unconditional_volatilities[ticker], h)
    garch_forecasts = compute_volatility_forecasts(garch_model, unconditional_volatilities[ticker], h)
    volatility_forecasts[ticker] = {'ARCH': arch_forecasts, 'GARCH': garch_forecasts}

# Plot volatility forecasts and unconditional volatility
for ticker in tickers:
    arch_forecasts = volatility_forecasts[ticker]['ARCH']
    garch_forecasts = volatility_forecasts[ticker]['GARCH']
    unconditional_volatility = unconditional_volatilities[ticker]

    plt.figure(figsize=(12, 6))
    plt.plot(range(1, h + 1), arch_forecasts, label='ARCH(3)', color='blue')
    plt.plot(range(1, h + 1), garch_forecasts, label='GARCH(1,1)', color='red')
    plt.axhline(y=unconditional_volatility, linestyle='--', color='green', label='Unconditional Volatility')
    plt.title(f'Volatility Forecasts for {ticker}')
    plt.xlabel('Number of Days Ahead')
    plt.ylabel('Volatility')
    plt.legend()
    plt.grid(True)
    plt.show()
```
