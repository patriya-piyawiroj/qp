```python
import yfinance as yf
import pandas as pd
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Download historical data
tickers = ['VFINX', 'VEURX', 'VEIEX', 'BLV', 'VBISX']
data = yf.download(tickers, start="2013-01-31", end="2020-12-31")['Close']
data = data.dropna()
returns = np.log(returns / returns.shift())

# Add edhec returns
edhec_data = pd.read_csv('edhec.csv', index_col='date', parse_dates=True)
equity_market_neutral = edhec_data['Equity.Market.Neutral']
data['Equity Market Neutral'] = equity_market_neutral

expected_returns = returns.mean()
cov_matrix = returns.cov()

# Step 6: Define the objective function (portfolio variance)
def portfolio_variance(weights, cov_matrix):
    return np.dot(weights.T, np.dot(cov_matrix, weights))

# Step 7: Constraints and bounds
constraints = ({'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1})
bounds = tuple((0, 1) for _ in range(len(expected_returns)))

# Initial guess
initial_guess = len(expected_returns) * [1. / len(expected_returns)]

# Step 8: Optimization
result = minimize(portfolio_variance, initial_guess, args=(cov_matrix,), method='SLSQP', bounds=bounds, constraints=constraints)

# Optimal weights
min_var_weights = result.x

# Step 9: Check for negative weights
negative_weights = any(w < 0 for w in min_var_weights)

# Step 10: Plot the weights
plt.figure(figsize=(10, 6))
plt.bar(returns.columns, min_var_weights)
plt.xlabel('Assets')
plt.ylabel('Portfolio Weights')
plt.title('Global Minimum Variance Portfolio Weights')
plt.show()

# Print results
print("Global Minimum Variance Portfolio Weights:")
print(pd.Series(min_var_weights, index=returns.columns))
print("\nAre there any negative weights?:", negative_weights)

portfolio_avg_return = np.sum(min_var_weights * expected_returns)
portfolio_std_dev = np.sqrt(np.dot(min_var_weights.T, np.dot(cov_matrix, min_var_weights)))

print("Average return of the portfolio:", portfolio_avg_return)
print("Standard deviation of the portfolio:", portfolio_std_dev)
```
